# Research Question — Lane 1: Ranking Signal Analysis

## Research Question
Which pages are gaining search visibility (rising impressions) without a
matching rise in clicks, signaling a missed opportunity to capture that
visibility?

## Unit of Analysis
One row = one page's search performance.

## Output
A ranked list of pages where impressions are rising but CTR/clicks
aren't keeping pace, sorted with the biggest "missed opportunity" pages
at the top — possibly supported by a chart.

## Action
For pages near the top of the list, a content/SEO team member reviews
the page and adds a clearer call-to-action and product link, to help
convert existing visibility into clicks/engagement.

## Cost of Being Wrong
Wasted team effort on pages that didn't need attention, and more
seriously real declining pages get overlooked while attention is
misdirected elsewhere.
